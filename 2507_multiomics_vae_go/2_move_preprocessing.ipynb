# moveの前処理

```python
import pandas as pd
```

```python
rna = pd.read_csv("data/rna_z_comsam.tsv", index_col=0, sep="\t", engine="python")
rna
```

```python
rna.index.name = None
```

```python
rna = rna.T
rna
```

```python
meth = pd.read_csv("data/meth_z_comsam.tsv",
                   index_col=0,
                   sep="\t",
                   engine="python")
meth
```

```python
meth = meth.T
meth
```

```python
cnv = pd.read_csv("data/cnv_z_comsam.tsv",
                   index_col=0,
                   sep="\t",
                   engine="python")
cnv
```

```python
cnv = cnv.T
cnv
```

```python
common = (rna.index.intersection(meth.index).intersection(cnv.index))
rna, meth, cnv = rna.loc[common], meth.loc[common], cnv.loc[common]
print(rna.shape, meth.shape, cnv.shape)
```

```python
all_ids = rna.index
```

```python
pd.Series(sorted(all_ids)).to_csv("data/com_sample.txt",index=False, header=False)
```

```python
rna.to_csv("data/rna_z_comsam_row.tsv", index=True, sep="\t")
meth.to_csv("data/meth_z_comsam_row.tsv", index=True, sep="\t")
cnv.to_csv("data/cnv_z_comsam_row.tsv", index=True, sep="\t")
```
